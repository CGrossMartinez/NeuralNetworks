```python
#variable declaration and initialization
positiveCount = 0
positiveCountNumerator = 0
negativeCount = 0
negativeCountNumerator = 0

#creating tuple with true/false values to compare
compareList = (df['sentiment'] == 'positive')

#first outter loop which traverse indexes of table
for i in range(0, len(df)):

  #vairiable declaration and initialization
  wordCount = 0

  #conditional statement which check if review is positive
  if str(compareList[i]) == 'True':

    #variable to keep track of number of positive reviews
    positiveCount += 1

    #inner loop which counts the number of words per review  
    for j in range(0, len(df.iloc[i,0])):

      #condition to ignore words with code formatting, e.g ,<br />
      if df.iloc[i, 0][j] == ' ' and str(df.iloc[i, 0][j + 1]) == '/':
        wordCount = wordCount
      
      #condition that counts one to wordCount when encountered by a 
      #space in string which denotes the end of a word 
      if str(df.iloc[i, 0][j]) == ' ':        
        wordCount += 1

    #updating variable with actual number of words per review to calculate avg
    positiveCountNumerator += wordCount

  #conditional statement which check if review is negative
  if str(compareList[i]) == 'False':

    #variable to keep track of number of negative reviews
    negativeCount += 1

    #inner loop which counts the number of words per review 
    for j in range(0, len(df.iloc[i,0])):

      #condition to ignore words with code formatting, e.g ,<br />
      if df.iloc[i, 0][j] == ' ' and str(df.iloc[i, 0][j + 1]) == '/':
        wordCount = wordCount
      
      #condition that counts one to wordCount when encountered by a 
      #space in string which denotes the end of a word
      if str(df.iloc[i, 0][j]) == ' ':        
        wordCount += 1
  
    #updating variable with actual number of words per review to calculate avg
    negativeCountNumerator += wordCount

#calculating and printing the avgs for each class of review to the screen
print('Average number of words in postive reviews =', str(positiveCountNumerator / positiveCount))
print('Average number of words in negative reviews =', str(negativeCountNumerator / negativeCount))
```

```python
#list declaration
wordListPositive = []
wordCounterListPositive = []
wordListNegative = []
wordCounterListNegative = []

#creating tuple with true/false values to compare
compareList = (df['sentiment'] == 'positive')

#first outter loop which traverse indexes of table
for i in range(0, len(df)):

  #variable declaration and initialization
  startIndex = 0
  endIndex = 0
  foundWord = 0

  #statement that checks if review is positive
  if str(compareList[i]) == 'True':

    #Inner loop which adds words to list and updates word count 
    for j in range(0, len(df.iloc[i, 0])):

      #loops which find end of word and saves index value
      while df.iloc[i, 0][endIndex] != ' ':        
        endIndex += 1   
     
      #statement that checks if list is empty
      if str(bool(wordListPositive)) == 'False':

        #code that appends word to word list and update the word count
        wordListPositive.append(df.iloc[i,0][startIndex : endIndex])      
        wordCounterListPositive.append(1)

        #updating the new index values for next iteration
        j = endIndex
        endIndex += 1
        startIndex = endIndex

      #statement that ensure wprd list is not empty
      else:
        
        #loop traversing word list
        for k in range(0, len(wordListPositive)):

          #statement that checks is the a is in the list
          #If word is on the list, count cvalue is updated 
          if str(wordListPositive[k]) == str(df.iloc[i, 0][startIndex:endIndex]):

            #updates the word count in the word count list            
            wordCounterListPositive[k] += 1
            
            #variable is updated to ensure word is not added to list again
            foundWord = 1

            #updating the new index values for next iteration
            j = endIndex
            endIndex =+ 1
            startIndex = endIndex
            #continue

        #condition that checks in word needs to be added to word list
        if foundWord != 1:
          
          #code that appends word to word list and update the word count
          wordListPositive.append(df.iloc[i,0][startIndex : endIndex])      
          wordCounterListPositive.append(1)

          #updating the new index values for next iteration
          j = endIndex
          endIndex += 1
          startIndex = endIndex

  #statement that checks if review is false
  #if review is false, code follows same structure as when it was true with  
  #the difference that the proper list for negative reiews will be updated
  if str(compareList[i]) == 'False':

    for j in range(0, len(df.iloc[i, 0])):
      
      while df.iloc[i, 0][endIndex] != ' ':        
        endIndex += 1   
     
      if str(bool(wordListNegative)) == 'False':

        wordListNegative.append(df.iloc[i,0][startIndex : endIndex])      
        wordCounterListNegative.append(1)

        j = endIndex
        endIndex += 1
        startIndex = endIndex        

      else:

        for k in range(0, len(wordListNegative)):

          if str(wordListNegative[k]) == str(df.iloc[i, 0][startIndex: endIndex]):

            wordCounterListNegative[k] += 1
            foundWord = 1

            j = endIndex
            endIndex =+ 1
            startIndex = endIndex
            #continue
          
        if foundWord != 1:
          
          wordListNegative.append(df.iloc[i,0][startIndex : endIndex])      
          wordCounterListNegative.append(1)

          j = endIndex
          endIndex += 1
          startIndex = endIndex

#printing lists with word and word count for the two types of reviews
print(wordListPositive)
print(wordCounterListPositive)
print(wordListNegative)
print(wordCounterListNegative)
```
